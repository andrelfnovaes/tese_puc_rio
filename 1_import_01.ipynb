```python
import yfinance as yf
import pandas as pd

# Import data

#The M6 asset universe
tickers = [
  "ABBV","ACN","AEP","AIZ","ALLE","AMAT","AMP","AMZN","AVB","AVY",
  "AXP","BDX","BF-B","BMY","BR","CARR","CDW","CE","CHTR","CNC",
  "CNP","COP","CTAS","CZR","DG","DPZ","DRE","DXC","META","FTV",
  "GOOG","GPC","HIG","HST","JPM","KR","OGN","PG","PPL","PRU",
  "PYPL","RE","ROL","ROST","UNH","URI","V","VRSK","WRK","XOM",
  "IVV","IWM","EWU","EWG","EWL","EWQ","IEUS","EWJ","EWT","MCHI",
  "INDA","EWY","EWA","EWH","EWZ","EWC","IEMG","LQD","HYG","SHY",
  "IEF","TLT","SEGA.L","IEAA.L","HIGH.L","JPEA.L","IAU","SLV","GSG","REET",
  "ICLN","IXN","IGF","IUVL.L","IUMO.L","SPMV.L","IEVL.L","IEFM.L","MVEU.L","XLK",
  "XLF","XLV","XLE","XLY","XLI","XLC","XLU","XLP","XLB","VXX"]

# Define begin_date and end_date
begin_date = '2022-02-01' 
end_date = '2023-02-28'

# Download historical data
data = yf.download(tickers, start=begin_date, end=end_date)

# Select only the adjusted close prices
data = data['Adj Close']

# Convert the index to datetime
data.index = pd.to_datetime(data.index)

display(data)

display("Data preprocessing completed.")
```

```python
# Print NaNs per column
for asset in data.columns:
    nan_count_per_column = data[asset].isna().sum()
    print(nan_count_per_column)
```

```python
# Drop columns entirely composed of NaNs
data.dropna(axis=1, how='all', inplace=True)

# Drop rows with NaNs
data.dropna(axis=0, how='any', inplace=True)

data
```

```python
# Print NaNs per column
for asset in data.columns:
    nan_count_per_column = data[asset].isna().sum()
    print(nan_count_per_column)
```

```python
# Save the 'data' DataFrame to a CSV file
data.to_csv('data_original_m6.csv', index=True)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define the ticker symbol
tickerSymbol = 'IBM'

# Adjust the date range
begin_date = '2024-03-29'
end_date = '2024-05-29'

# Get the data
tickerData = yf.download(tickerSymbol, start=begin_date, end=end_date)

# Select the close prices
close_prices = tickerData['Close']

# Create a new figure with a specified size (width, height)
plt.figure(figsize=(10, 6))

# Create a scatter plot with larger points
plt.scatter(close_prices.index, close_prices, color='blue', s=100)

# Add a line plot
plt.plot(close_prices.index, close_prices, color='blue')

# Change the x-axis labels to a smaller size and simplified format
plt.xticks(fontsize=8, rotation=45)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Add extra space on the right side of the plot
plt.xlim([datetime.strptime(begin_date, '%Y-%m-%d'), datetime.strptime(end_date, '%Y-%m-%d') + timedelta(days=30)])

plt.title('IBM Close Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
